#### Exercice 1 : Création de graphiques
**Contexte :** Lors de l'analyse de données expérimentales en physique, il est courant de devoir visualiser des fonctions mathématiques. Dans cet exercice, nous allons générer un graphique pour représenter la fonction sin(x) sur un intervalle donné. Cette compétence est précieuse pour représenter graphiquement des phénomènes périodiques en physique.

- Générez un tableau NumPy représentant la fonction sin(x) sur l'intervalle [0, 2π].
- Utilisez Matplotlib pour créer un graphique de la fonction.
- Assurez-vous d'ajouter des étiquettes appropriées pour les axes et un titre au graphique.

```python
import numpy as np
import matplotlib.pyplot as plt

# Générer un tableau NumPy représentant sin(x) sur l'intervalle [0, 2π]
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Utiliser Matplotlib pour créer un graphique
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Graphique de sin(x) sur [0, 2π]')
plt.show()
```

#### Exercice 2 : Statistiques des données
**Contexte :** En physique, nous effectuons souvent des mesures expérimentales et nous devons comprendre les propriétés statistiques des données que nous avons collectées. Cet exercice vous permettra de calculer des statistiques de base et de les représenter graphiquement.

- Utilisez NumPy pour générer un ensemble de données fictives, par exemple, des mesures expérimentales.
- Calculez la moyenne et l'écart type de ces données.
- Utilisez Matplotlib pour créer un histogramme de ces données avec des étiquettes appropriées pour l'axe x et le titre.


La densité de probabilité (PDF, Probability Density Function en anglais) d'une variable aléatoire continue est une fonction qui décrit la probabilité de la variable aléatoire de prendre une valeur particulière. Pour une variable aléatoire continue X, la densité de probabilité f(x) est définie mathématiquement comme suit :

- $f(x) \geq 0$  $\forall$ $ x $

- $\int_{-\infty}^{\infty} f(x)$ , $dx = 1$ 

En d'autres termes, la densité de probabilité est une fonction non négative telle que l'intégrale de cette fonction sur l'ensemble de toutes les valeurs possibles de la variable aléatoire est égale à 1. Cela signifie que la probabilité totale que la variable aléatoire prenne n'importe quelle valeur dans son ensemble de valeurs possibles est égale à 1.

La forme précise de la densité de probabilité dépend de la distribution de probabilité spécifique qui décrit la variable aléatoire. Par exemple, pour une distribution normale (gaussienne), la densité de probabilité est donnée par la célèbre courbe en forme de cloche :

\begin{equation}
f(x) = \frac{1}{\sigma\sqrt{2\pi}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right) 
\end{equation}

où :
- $x$ est la valeur à évaluer.
- $\mu $ est la moyenne de la distribution.
- $ \sigma $ est l'écart type de la distribution.
- $ \pi $ est la constante mathématique.

Pour d'autres distributions de probabilité, la forme de la densité de probabilité sera différente. Chaque distribution a sa propre formule mathématique spécifique pour sa densité de probabilité.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  # Importer la distribution normale

# Générer un ensemble de données fictives
data = np.random.normal(0, 1, 1000)  # Exemple de données normales

# Calculer la moyenne et l'écart type
mean = np.mean(data)
std = np.std(data)

# Créer un histogramme des données
plt.hist(data, bins=30, density=True, alpha=0.5, color='b', label='Données')
plt.xlabel('Valeurs')
plt.ylabel('Fréquence')
plt.title('Histogramme de données')

# Ajustement gaussien
x = np.linspace(-4, 4, 100)
p = norm.pdf(x, mean, std)  # Distribution gaussienne correspondante
plt.plot(x, p, 'k', linewidth=2, label='Ajustement Gaussien')

plt.legend()
plt.show()

# Afficher la moyenne et l'écart type
print(f'Moyenne : {mean}')
print(f'Écart type : {std}')
```

#### Exercice 3 : Régression linéaire
**Contexte :** La régression linéaire est un outil puissant en physique pour modéliser des relations entre des variables. Dans cet exercice, nous allons utiliser NumPy et Matplotlib pour effectuer une régression linéaire simple.

- Donnez un ensemble de données x et y.
- Utilisez NumPy pour calculer une régression linéaire sur ces données.
- Tracez à la fois le nuage de points des données et la ligne de régression linéaire à l'aide de Matplotlib. Assurez-vous d'ajouter des étiquettes pour les axes et une légende.


```python
# Données d'exemple
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 3.5, 5, 6])

# Calcul de la régression linéaire ( x,y, degrés du polynôme)
coefficients = np.polyfit(x, y, 1)
a, b = coefficients

# Création du graphique
plt.scatter(x, y, label='Données')
plt.plot(x, a * x + b, 'r', label='Régression linéaire')
plt.xlabel('x')
plt.ylabel('y')

# Affichage de l'équation de la régression linéaire
equation = f'Régression linéaire : y = {a:.2f}x + {b:.2f}'
plt.text(1, 4.9, equation, fontsize=12, color='r')

plt.title('Régression linéaire')
plt.legend()
plt.show()
```

#### Exercice 4 : Transformation de Fourier
**Contexte :** La transformation de Fourier est un outil essentiel pour analyser les signaux périodiques en physique. Cet exercice vous guidera pour effectuer une transformation de Fourier sur un signal discret et afficher son spectre de fréquence.

- Fournissez un signal discret sous forme d'array NumPy (par exemple, une somme de sinusoïdes).
- Appliquez la transformation de Fourier à ces données à l'aide de NumPy.
- Utilisez Matplotlib pour tracer le spectre de fréquence du signal. Assurez-vous d'ajouter des étiquettes pour les axes et un titre au graphique.

```python
import numpy as np
import matplotlib.pyplot as plt

# Exemple de signal (somme de sinusoïdes)
t = np.linspace(0, 1, 1000)
frequency = 50
frequency2 = 100
signal = np.sin(2 * np.pi * frequency * t) + 0.5 * np.sin(4 * np.pi * frequency2 * t)+ np.cos(150 * np.pi * frequency * t)

# Calcul de la transformation de Fourier
fourier_transform = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(t))
positive_frequencies = frequencies[:len(frequencies) // 2]
spectrum = np.abs(fourier_transform)[:len(t) // 2]
# Creation du graphique spatial 

plt.plot(t, signal)
plt.xlabel('Temps (s)')
plt.ylabel('Amplitude')
plt.title('Signal temporel')
plt.show()

# Création du graphique
plt.plot(positive_frequencies, spectrum)
plt.xlabel('Fréquence (Hz)')
plt.ylabel('Amplitude')
plt.title('Spectre de fréquence')
plt.show()
```

#### Exercice 5 : Simulation de Monte Carlo pour l'estimation de π
**Contexte :** La méthode de Monte Carlo est utilisée pour estimer des valeurs numériques en simulant des processus aléatoires. Dans ce cas, nous allons utiliser cette méthode pour estimer la valeur de π en générant des points aléatoires.

- Générez un grand nombre de points aléatoires dans un carré de côté 2, dont un quart de cercle est inscrit.
- Utilisez ces résultats pour estimer la valeur de π et créez un graphique montrant à la fois les points aléatoires et le quart de cercle. Ajoutez une légende et assurez-vous que le graphique est à l'échelle.

```python
# Générer un grand nombre de points aléatoires dans un carré de côté [0, 1] x [0, 1]
n = 10000
points = np.random.rand(n, 2)
inside = 0
for i in range(n):
    x, y = points[i]
    if x**2 + y**2 <= 1:
        plt.plot(x, y, 'b.')
        inside += 1
    else:
        plt.plot(x, y, 'r.')
pi = 4 * inside / n
plt.title(f'Pi = {pi:.10f}')
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Générer un grand nombre de points aléatoires dans le carré [0, 1] x [0, 1]
n_points = 10000
points = np.random.rand(n_points, 2)

# Vérifier si les points sont à l'intérieur du quart de cercle
inside_quarter_circle = np.sum(points[:, 0]**2 + points[:, 1]**2 <= 1)

# Estimation de π
pi_estimate = 4 * inside_quarter_circle / n_points

x = np.linspace(0, 1, 100)
y = np.sqrt(1 - x**2)

# Création du graphique
plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], s=1, label='Points aléatoires')
plt.plot(x, y, 'r', label='Cercle')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Estimation de π ≈ {pi_estimate:.10f}')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
# Générer un grand nombre de points aléatoires dans le carré [0, 1] x [0, 1]
n_points = 100000
points = np.random.rand(n_points, 2)

# Calculer la distance de chaque point par rapport à l'origine
distances = np.sqrt(points[:, 0]**2 + points[:, 1]**2)

# Créer un masque pour les points à l'intérieur du quart de cercle
inside_mask = distances <= 1

# Utiliser le masque pour diviser les points en deux ensembles
inside_points = points[inside_mask]
outside_points = points[~inside_mask]

# Estimation de π
pi_estimate = 4 * len(inside_points) / n_points

x = np.linspace(0, 1, 100)
y = np.sqrt(1 - x**2)

# Création du graphique
plt.figure(figsize=(6, 6))
plt.scatter(inside_points[:, 0], inside_points[:, 1], s=1, c='r', label='Points à l\'intérieur')
plt.scatter(outside_points[:, 0], outside_points[:, 1], s=1, c='b', label='Points à l\'extérieur')
plt.plot(x, y, 'green', label='Cercle')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Estimation de π ≈ {pi_estimate:.10f}')
plt.legend()
plt.axis('equal')
plt.show()
```
