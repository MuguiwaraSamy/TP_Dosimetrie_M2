**Question 8**  
Comment créez-vous un tableau NumPy de valeurs aléatoires entre 0 et 1 ?

```python
import matplotlib.pyplot as plt

normal = np.random.normal(0,1,1000)
plt.hist(normal, bins=100)
plt.show()

uniform = np.random.uniform(0,1,1000)
plt.hist(uniform, bins=100)
plt.show()

poisson = np.random.poisson(1,1000)
plt.hist(poisson, bins=100)
plt.show()
```
